```python
import yfinance as yf
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
```

```python
# set start and end dates
start_date = '2022-02-08'
end_date = '2023-02-08'

# ticker
ticker = 'AMZN'

# get data
data = yf.download(ticker, start_date, end_date)

# adjust df for plotting
data['Date'] = data.index
data = data[['Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']]
data.reset_index(drop=True, inplace=True)
```

```python
np.arange(len(data.index))
```

```python
data['index'] = np.arange(len(data.index))
```

```python
data.head()
```

```python
# fig, ax = plt.subplots()
fig, ax = plt.subplots(figsize=(16,9))
img = plt.imread('../static/amazon_alexa.jpg')
plt.plot(data['index'], data['Open'], linewidth = 20, color='red', alpha=0.3)
fig.figimage(img, 0, 0, resize=True, zorder=-1)

plt.tight_layout()
plt.axis('off')
plt.margins(x=0)
plt.savefig("output.jpg")
plt.show()
```
